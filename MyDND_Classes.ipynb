```python
import enum
import random

def RollTheDice(k):
    # бросок кубика с k гранями
    return int(random.uniform(1,k+1))

class Monster():
    def __init__(
        self, kind, name, hp, kd, weapon, attack_bonus,
        damage_dice, damage_bonus):
        self.kind = kind
        self.name = name
        self.HP = hp
        self.kd = kd
        self.weapon = weapon
        self.attack_bonus = attack_bonus
        self.damage_dice = damage_dice
        self.damage_bonus = damage_bonus

    def InfoM(self):
        # информация о параметрах монстра
        print(f'{self.kind} {self.name}')
        print(f'Здоровье (ХП): {self.HP}. Класс доспехов: {self.kd}')
        print(f'Бонус атаки {self.attack_bonus}')
        print(f'Оружие {self.weapon}. Кость урона 1к{self.damage_dice}, бонус урона {self.damage_bonus}')

    def FightAttack(self, other):
        # метод реализует нападение монстра на перонажа
        rolldice20 = RollTheDice(20)
        print("Атакует",self.kind,self.name,". На кубике выпало",rolldice20,
        "+ бонус атаки",self.attack_bonus,"итого",
        rolldice20 + self.attack_bonus)
        print("Атакованный", other.nameH, "облачен в доспехи уровня",
              other.kd)
        if rolldice20 + self.attack_bonus >= other.kd:
            damageRoll = RollTheDice(self.damage_dice)
            print("Успешная атака наносит", damageRoll, "+", self.damage_bonus,
                  "=", damageRoll + self.damage_bonus, "урона")
            other.Injury(damageRoll + self.damage_bonus)
        else:
            print(other.nameH, " не получает урон от этой атаки")

    def Injury(self,damage):
        self.HP = max(0,self.HP - damage)
        if self.HP == 0:
            print(self.name, " уничтожен. Опишите как вы это сделали.")

class RaceH(enum.Enum):
    # возможные расы персонажей,
    Elf = 'Elf'
    Human = 'Human'
    Dwarf = 'Dwarf'
    Halfling = 'Halfling'
    Gnome = 'Gnome'
    Dragonborn = 'Dragonborn'

class ClassH(enum.Enum):
    # возможные классы персонажей
    Barbarian = 'Barbarian'
    Warrior = 'Warrior'
    Rogue = 'Rogue'
    Bard = 'Rogue'
    Cleric = 'Cleric'


class DnDHero:

    def __init__(self, nameH, raceHero : RaceH, classHero : ClassH, MaxHP, kd,
                 weapon, attack_bonus,damage_dice, damage_bonus, age : int):
        # обязательные параметры передаются в конструктор
        # без них нельзя создать новый объект - персонажа
        # раса персонажа не может измениться
        self.nameH = nameH
        self.__race__ = raceHero
        self.experience = 0 #у нового персонажа опыт равен нулю
        self.level = 1 #новый персонаж 1 уровня
        self.classHero = classHero
        self.MaxHP = MaxHP
        self.HP = self.MaxHP
        self.kd = kd
        self.weapon = weapon
        self.attack_bonus = attack_bonus
        self.damage_dice = damage_dice
        self.damage_bonus = damage_bonus
        self.age = age
        # необязательные параметры вновь созданного персонажа.
        # могут быть присвоены позже
        # Базовые характеристики персонажа
        # [0] русское название характеристики
        # [1] постоянное значение характиристики на текущем уровне
        # [2] временные изменение характеристики в текущем состоянии персонажа:
        # "0" - нет штафов и бонусов; "+" бонус к характеристики; "-" штраф
        self._baseParam = {
            "strength": ["Сила", None, None],
            "agility": ["Ловкость", None, None],
            "constitution": ["Телосложение", None, None],
            "intelligence": ["Интеллект", None, None],
            "wisdom": ["Мудрость", None, None],
            "charisma": ["Харизма", None, None]}

    def SetBaseParamRND(self):
        # генерация случайных значений характеристик персонажа
        for key in self._baseParam:
            self._baseParam[key][1] = int(random.uniform(8,14))

    def PrintBaseParam(self):
        print(
            "У персонажа ", self.nameH, " на уровне ",
            self.level ," следующие характеристики"
            )
        for key in self._baseParam:
            print(self._baseParam[key][0], " = " , self._baseParam[key][1])

    def FullInfo(self):
        print("Вашего персонажа зовут", self.nameH)
        print("Раса", self.__race__.value)
        print("Класс", self.classHero.value)
        print("Уровень", self.level)
        print("Опыт", self.experience)
        print("Возраст", self.age)
        print("Максимальное здоровье (ХП):", self.MaxHP,
              ", текущее здоровье", self.HP)
        print("Класс доспехов: ", self.kd)
        print("Бонус атаки", self.attack_bonus)
        print("Оружие ", self.weapon, ". Кость урона 1к",
              self.damage_dice, " бонус урона ", self.damage_bonus,sep="")
        self.PrintBaseParam()

    def Injury(self,damage):
        self.HP = max(0,self.HP - damage)
        if self.HP == 0:
            print(self.nameH, "критически ранен и находится при смерти.",
                  "Необходимо выполнить спасброски от смерти")

    def FightAttack(self, other):
      # метод реализует атаку персонажа
        rolldice20 = RollTheDice(20)
        print("Атакует",self.nameH,". На кубике выпало",rolldice20,
        "+ бонус атаки",self.attack_bonus,"итого",
        rolldice20 + self.attack_bonus)
        print("Атакованный", other.kind, other.name , "облачен в доспехи уровня",
              other.kd)
        if rolldice20 + self.attack_bonus >= other.kd:
            damageRoll = RollTheDice(self.damage_dice)
            print("Успешная атака наносит", damageRoll, "+", self.damage_bonus,
                  "=", damageRoll + self.damage_bonus, "урона")
            other.Injury(damageRoll + self.damage_bonus)
        else:
            print(other.kind, other.name , " не получает урон от этой атаки")

# создадим парочку монстров
M1 = Monster("Зомби", "Мертвый Вася", 6, 12, "Короткий меч", 4, 6, 2)
M1.InfoM()
M2 = Monster("Дракон", "Смауг", 200, 29, "Огненное дыхание", 14, 12, 8)
M2.InfoM()

# и одного геройчика
Pers1 = DnDHero("Оскар", RaceH.Elf,ClassH.Warrior, 18, 16, "Длинный лук",
                6, 8, 4, 177)
Pers1.SetBaseParamRND()

Pers1.FullInfo()

# 3 2 1 ... Fight!
print("Боевая разминка")
M1.FightAttack(Pers1)
Pers1.FightAttack(M1)
M2.FightAttack(Pers1)
if Pers1.HP > 0:
    print(f'У персонажа {Pers1.nameH} осталось {Pers1.HP} ХП')
```
